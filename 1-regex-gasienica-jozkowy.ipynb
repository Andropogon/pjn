```python
import os
import regex
import pandas as pd
import matplotlib.pyplot as plt
```

```python
addition_dict = {}
removal_dict = {}
change_dict = {}
```

```python
directory = r'./ustawy'
for filename in os.listdir(directory):
    with open(directory +'/'+filename , encoding='utf-8') as f:
        file_content = f.read().replace("\n", " ")
        additions = len(regex.findall('dodaje się (art.|ust.|pkt|rozdział|nowy|§)', file_content))
        deletions = len(regex.findall("(skreśla się( art.| ust.|:| lit| pkt| w ust.|.| §| poz.))|(uchyla się( ust.| art.| pkt| zdanie| §))", file_content)) 
        changes = len(regex.findall("otrzymuj(e|ą) brzmienie", file_content)) 
        addition_dict[filename] = additions
        removal_dict[filename] = deletions
        change_dict[filename] = changes
```

```python
from collections import defaultdict
act_in_years = defaultdict(lambda: [0,0,0])
```

```python
# 1,2 - For each bill compute the number of the following amendments present in the bill
import operator
for key in addition_dict:
    print('Additions:', key, addition_dict[key])
    print('Removal:', key, removal_dict[key])
    print('Changes:', key, change_dict[key])
    print('--- ---- ----')
    act_in_years[key.split('_')[0]] = list(map(operator.add, act_in_years[key.split('_')[0]],[addition_dict[key], removal_dict[key], change_dict[key]]))
```

```python
act_in_years
```

```python
# 3 - Plot results from point 1 showing how the percentage of amendments of a given type changed in the consecutive years.
act_in_years_chart = act_in_years
for x in act_in_years:
    for y in range(len(act_in_years[x])):
        act_in_years_chart[x][y] = act_in_years[x][y]/sum(act_in_years[x])
df = pd.DataFrame.from_dict(act_in_years_chart, orient = 'index')
df = df.rename(columns = {0: 'Additions', 1: 'Deletions', 2: 'Changes'})
df.plot(x=df.index.name, y = ['Additions', 'Deletions', 'Changes'], kind="bar")
```

```python
# 4 - Compute the total number of occurrences of the word ustawa in any inflectional form (ustawa, ustawie, ustawę, etc.) and all spelling forms (ustawa, Ustawa, USTAWA), excluding other words with the same prefix (e.g. ustawić).
# 5 - Compute the total number of occurrences of the same word (same conditions), followed by z dnia expression.
# 6 - As above, but not followed by z dnia expression. Is the result correct (result 4 =? result 5 + result 6)?
# 7 - Compute the total number of occurrences of the word ustawa in any inflectional form, excluding occurrences following o zmianie expression.
dict_ustawa = {}
dict_ustawa_z_dnia = {}
dict_not_ustawa_z_dnia = {}
dict_ustawa_not_o_zmianie = {}
for filename in os.listdir(r'./ustawy'):
    with open(directory +'/'+filename , encoding='utf-8') as f:
        file_content = f.read().replace("\n", " ")
        findall_ustawa = regex.findall(r'\bustaw((\b)|(a|y|ie|ę|ą|o|om|ami|ach)\b)', file_content, regex.IGNORECASE)
        for i in findall_ustawa:
            if i[0].isupper():
                key = 'USTAW'+ i[0]
            else:
                key ='ustaw'+ i[0]
            if key in dict_ustawa.keys():
                dict_ustawa[key]+=1
            else:
                dict_ustawa[key]=1
                
        for key, item in dict_ustawa.items():
            findall_z_dnia_counter = len(regex.findall(r'\b{0}+ z dnia \b'.format(key), file_content))
            if key in dict_ustawa_z_dnia.keys():
                dict_ustawa_z_dnia[key] += findall_z_dnia_counter
            else:
                dict_ustawa_z_dnia[key] = findall_z_dnia_counter
                
            # I had some problem to count this senteces, results are slighty wrong
            findall_not_z_dnia_counter = len(regex.findall(r'\b{0}(?! z dnia)\b'.format(key), file_content))
            if key in dict_not_ustawa_z_dnia.keys():
                dict_not_ustawa_z_dnia[key] += findall_not_z_dnia_counter
            else:
                dict_not_ustawa_z_dnia[key] = findall_not_z_dnia_counter

        for key, item in dict_ustawa.items():
            findall_ustawa_not_o_zmianie_counter = len(regex.findall(r'\b{0}(?! o zmianie)\b'.format(key), file_content))
            if key in dict_ustawa_not_o_zmianie.keys():
                dict_ustawa_not_o_zmianie[key] += findall_ustawa_not_o_zmianie_counter
            else:
                dict_ustawa_not_o_zmianie[key] = findall_ustawa_not_o_zmianie_counter
```
